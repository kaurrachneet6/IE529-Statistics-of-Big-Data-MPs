# Appendix 1: Python Script to plot the Scatter Plots of the given datasets

```python
import numpy as np
import matplotlib.pyplot as plt
```

$\textbf{Reading the datasets}$

```python
X = np.genfromtxt(input(),delimiter = ',', dtype=None) 
# The file name for data matrix as an input from the user (Eg: Dataset 1/clustering.csv)
```

```python
Y = np.genfromtxt(input(),delimiter = ',', dtype=None) 
# The file name for data matrix as an input from the user (Eg: Dataset_2/ShapedData.csv)
```

$\textbf{Scatter Plot for the dataset 1 - clustering.csv}$

```python
#Dataset 1 - clustering.csv
plt.figure(figsize=(10,6))
plt.plot(X[:,0],X[:,1], 'ob')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Scatter plot for Dataset 1 - clustering.csv')
plt.show()
```

$\textbf{Scatter Plot for the dataset 2 - ShapedData.csv}$

```python
#Dataset 2 - ShapedData.csv
plt.figure(figsize=(10,6))
plt.plot(Y[:,0],Y[:,1], 'or')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Scatter plot for Dataset 2 - Shaped Data')
plt.show()
```
